```python
print("--- Simulasi Pengisian Tangki ---")

tangki = 0  # mulai dari 0 liter
kapasitas = 100  # batas maksimum tangki

while tangki <= kapasitas:
    print(f"Volume tangki sekarang: {tangki} liter")

    # Logika kondisi peringatan
    if tangki == 80:
        print("WARNING: Hampir penuh!")
    elif tangki == 100:
        print("TANGKI PENUH – Pengisian berhenti.")
        break  # keluar dari loop jika sudah penuh

    # Tambah 10 liter tiap iterasi
    tangki += 10
```

```python
# Program: Simulasi Perhitungan Energi Listrik Rumah Tangga
# Konsep: Menggunakan if, elif, else, for, while, dan fungsi (function)

print("=== SIMULASI PENGGUNAAN ENERGI LISTRIK ===")

# -------------------------------
# 1️⃣ FUNGSI UNTUK MENGHITUNG ENERGI
# -------------------------------
def hitung_energi(daya_watt, jam_pakai_per_hari):
    """
    Menghitung energi listrik (kWh) per bulan
    Rumus: Energi (kWh) = (Daya (W) × Jam × 30 hari) / 1000
    """
    energi = (daya_watt * jam_pakai_per_hari * 30) / 1000
    return energi


# -------------------------------
# 2️⃣ FUNGSI UNTUK MENENTUKAN TARIF LISTRIK
# -------------------------------
def hitung_tarif_listrik(kwh):
    """
    Menentukan biaya listrik berdasarkan pemakaian energi
    """
    if kwh <= 100:
        tarif = 1500
    elif kwh <= 300:
        tarif = 2000
    else:
        tarif = 2500
    return kwh * tarif


# -------------------------------
# 3️⃣ INPUT JUMLAH PERALATAN DENGAN VALIDASI (while loop)
# -------------------------------
while True:
    try:
        jumlah_alat = int(input("Masukkan jumlah peralatan listrik yang ingin dihitung: "))
        if jumlah_alat > 0:
            break
        else:
            print("Jumlah alat harus lebih dari 0!")
    except ValueError:
        print("Input tidak valid! Harap masukkan angka.")


# -------------------------------
# 4️⃣ MENGINPUT DATA PERALATAN (for loop)
# -------------------------------
total_kwh = 0
for i in range(1, jumlah_alat + 1):
    print(f"\nPeralatan ke-{i}:")
    nama = input("Nama alat listrik: ")
    daya = float(input("Daya alat (Watt): "))
    jam = float(input("Jam penggunaan per hari: "))

    energi = hitung_energi(daya, jam)
    total_kwh += energi

    print(f"Energi bulanan {nama}: {energi:.2f} kWh")


# -------------------------------
# 5️⃣ HITUNG TOTAL BIAYA DENGAN PEMANGGILAN FUNGSI
# -------------------------------
total_biaya = hitung_tarif_listrik(total_kwh)


# -------------------------------
# 6️⃣ OUTPUT HASIL DAN KLASIFIKASI PENGGUNAAN
# -------------------------------
print("\n=== HASIL PERHITUNGAN ===")
print(f"Total konsumsi energi: {total_kwh:.2f} kWh per bulan")
print(f"Total biaya listrik: Rp {total_biaya:,.0f}")

# Percabangan logika
if total_kwh <= 100:
    print("Kategori: Penggunaan Rendah (Hemat Energi)")
elif total_kwh <= 300:
    print("Kategori: Penggunaan Sedang")
else:
    print("Kategori: Penggunaan Tinggi (Perlu Penghematan!)")

print("\n=== Program Selesai ===")
```
